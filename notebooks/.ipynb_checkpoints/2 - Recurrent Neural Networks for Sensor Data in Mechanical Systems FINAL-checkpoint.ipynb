# Overview

Mechanical systems are often fitted with a slew of sensors to monitor real time performance and equipment health. This data is extremely valuable in identifying issues based on a set of rules and diangosing failures. Depending on the system, data can be collected in intervals of milliseconds to hours. Regardless, this information is extremely valuable and is primed for machine learning. 

This application of machine learning can be used to identify very complex systems based on a high order of inputs, predict future states of equipment all of which can prevent unplanned shutdowns or help schedule system maintenace. Using such models optimizes company resources resulting in cost savings.

This example highlights a bladder accumulator system which is often vital to rotating equipment and lubrication systems. A bladder, or ballon, is often filled with nitrogen and applies a pressure to a synthetic lubricant which overtime can aid in operation of bearings, mechanical seals, etc. Most often the pressure is monitored and a set of predefined rules are used to set alarms indicating the systems needs attention. This design methodology may not account for all the variables that influence the performance of system. Once installed on a piece of equipment, we can begin collecting and analyzing data *SPECIFIC* to that unique installation. A practical example is a bladder accumulator initially filled by an operator, the system pressure is monitored and the fluid is consumed over time. Once the pressure sensor reaches the low level alarm, the system is refilled and the cycle continues.

![Bladder Accumulator](../figs/BladderAccumulator.png)

What if we could use machine learning to predit future performance or a future state based on a significantly more complex model?

## The Problem

Lack of planning or insight on equipment health and performance often ends in abrupt failure and equipment downtime. This accrues excessive costs and can strain manpower and equipment in order to become operable again. Logistics of managing a high volume of equpipment can become extremely challenging and less than optimal.

There is often no model that evalutes external paramters such as ambient conditions or equipment performance for a unique installation. These evaluations are painted with a wide brush. 

## The Solution

By implementing machine learning in these systems, we can create complex models that are specific to the system. These models can provide future predictions on performance that can be used for proactive system maintence and resource optimization. With a high volume of data, we can ellaboratly model complex systems and predict future states to optimize utilization and efficiency.

# Project Setup

In this application, we'll imagine the bladder accumulator is installed in the desert of Southern California at a solar power plant (this is a very real situation). The accumulator is charged to a set pressure and the equipment operates for two years. Measurements are taken every 2 hours recording:

1. Bladder Pressure
2. Equipment Flow Rate
3. Ambient Temperature

The consumption of the synthetic lubricant is often dependent on the equipment operating speed. Rotations per minute (RPM) could also be used but we know that flow rate for a pump is dependent on the rotational speed. Generally a higher speed (or flow), results in higher consumption of the synthetic fluid. In addition, the ambient temperature in the high desert of Souther California can be extreme. The bladder accumulator is essentially a ballon. Ever woken up on a cold morning and the tire pressure sensor came on in your car? The ideal gas law says the same will happen to the bladder accumulator! These temperature extremes will create daily fluctuations in the bladder pressure which may or may not trigger an alarm. It can also trigger false alarms often because modern 'rule-based' systems don't account for such complex variables.

We can build a specific machine model that will account for all these factors in predicting future stats of the equipment.

# Import Time Series Data

The data was synthetically generated based on the yearly average temperatures in the CA desert. General pressure drop rates were determined based on a randomized flow rate. The equipment could operate at 0,5000, or 8000 GPM (gallons per minute). The pressure decay rate increased as a function of flow. We'll imagine the system was primed to 60 PSI and requires refill at 35 PSI. This data was generated in the first notebook and exported to a csv. This remainder of the notebook will follow the *main.py* file which contains the machine learning specifics. 

The generated data is a great approximation of what an actual system may be recording in the field.

```python
import pandas as pd
csv_path = "../data/sys_data.csv"
df = pd.read_csv(csv_path)
df.head(5)
```

The dataframe above shows data intitialized in the month of January. We can see the equipment is not operating at first when the system was primed to 60 PSI. Let's plot the time series data for review.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import os

time = range(len(df["Temp"]))

fig,axs = plt.subplots(3,sharex=True,figsize=(10,6))
fig.suptitle("Generated Data for Mechanical Support System")
axs[0].plot(time,df['Temp'],'tab:orange')
axs[0].set_ylabel("Ambient Temp")
axs[1].plot(time,df['Flow Rate'],'tab:green')
axs[1].set_ylabel("Flow Rate")
axs[2].plot(time,df["Pressure"])
axs[2].set_ylabel("System Pressure")

#fig.savefig(os.path.join('../figs',"RawData.png"))
```

This data is generated over a 2 year period with measurements every 2 hours. In the first plot, we see the yearly cycle of temperature variation along with the daily highs and lows. The second plot illustrates the flow rate which may depend on the plant demand. We'll see a clear correlation to pressure drop in the system when the pump is in 'standby' (i.e. zero flow). During standby, the pressure downward trend is minimal but we can see the variations within the daily high and low temperatures. There is limited consumption of the fluid through this period. We can also see when the equipment operates at 5000 GPM, the synthetic fluid takes longer to hit the low alarm versus when it operates at 8000 GPM. 

Let's look at the pressure of a course of 5 days. We see temperatures swing throughout the 5 days between 45-70 F which is typical for January. The equipment is initially in standby so we have very little pressure decay due to non-operation. We can see how the system pressure is in phase with the ambient temperature. The high alarm setting is never reached but if miscalcualted, the operator of the equipment could be fooled although the alaram was triggered by the temperature varitions. Same could be true for a low alarm setting.

```python
plt.figure(figsize=(10,4))
plt.plot(time[:60],df['Temp'][:60],'tab:orange')
plt.plot(time[:60],df['Pressure'][:60])
plt.ylabel("System Pressure [psi]")
plt.xlabel("Time")
plt.legend=(['Temp','Pressure'])
plt.title("System Pressure Over First 5 Days")
```

# Split Train and Validation 

Now that we have the raw data imported, we can pass it to a split function. We'll use 80% train and 20% validation for this example. This is generally a good rule of thumb. We can then see the last 1728 points will be used for validation of the model.

```python
# Import from project localized folder
import sys
sys.path.append('../src/')

from data import DataSplit, AvgNormalization
from model import CNN,RNN,compile_and_fit

split_index = 0.8
df_train, df_val = DataSplit(df,split_index)
```

# Normalizing the Data

Next we want to normalize the 3 data parameters before feeding it into our model. We'll use only the training data mean and standard deviation to remove bias towards the validation data. For each data point, we'll subtract out the average of the training data and divide it by the standard deviation.

```python
train_df, val_df = AvgNormalization(df_train, df_val)
```

# Creating Input Series and Labels

Next, each dataframe is split into sequences of user defined window size. In this stud, a label size of 1 will be used (i.e. the next sequential number) to label the input data. For example, if we used a window of 6 and the array was `[ 1,2,3,4,5,6 ]` the windowing function will take the array and split it into input data `[ 1,2,3,4,5 ]` and `[ 6 ]` being the label for that sequence. 

In the `src` directory, a class was created to handle this task so various window and label sizes can be fed into the models. We iniate the class as such and can print the initialized class, to see specifics returned on the windowed data.

```python
from windowdata import WindowGenerator

w1 = WindowGenerator(input_width=24,label_width=1,shift=1,train_df=train_df,
                    val_df=val_df,label_columns=['Pressure'])
```

```python
print(w1)
```

The sequences have been batched in sets of 32. The 'time' is simply the window size minus the specified label width. The features vary between the input and labels because we are only using this model to predict pressure. Therefore our labels have a feature size of 1. Our inputs however include the 3 sensor data:

1. Pressure
2. Flow Rate
3. Ambient Temperature
